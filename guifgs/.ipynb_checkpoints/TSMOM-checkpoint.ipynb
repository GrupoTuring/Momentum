```python
assets = list(final_data.columns)
```

```python
final_data.columns
```

```python
def sign(x):
    return 1 if x > 0 else -1
```

```python
def tsmom(twelve_month_data, one_month_data):
    asset_return = {}
    #asset_return_long_only = {}
    for asset in assets:
        tsmom_return = []
        #long_only_return = []
        current_date = twelve_month_data.iloc[252:].index[0]
        for date in twelve_month_data.index[253:]:
            if date.month != current_date.month: # mapeia transiÃ§Ã£o entre meses
                r = sign(twelve_month_data[asset][current_date]) * one_month_data[asset][current_date]
                #long_only = one_month_data[asset][current_date]
                tsmom_return.append(r)
                #long_only_return.append(long_only)
            current_date = date
        #sum(tsmom_return)
        asset_return[asset] = tsmom_return
        #asset_return_long_only[asset] = long_only_return
    return asset_return
```

```python
asset_return = tsmom(twelve_month_data, one_month_data)
```

```python
def accumulated_return(asset_return):
    asset_accumulated_return = {}
    for asset in asset_return:
        accumulated = []
        initial = 1
        for r in asset_return[asset]:
            initial *= (1 + r)
            accumulated.append(initial)
        asset_accumulated_return[asset] = accumulated
    return asset_accumulated_return
```

```python
#long_only_AN = accumulated_return(asset_return_long_only)
tsmom_accumulated_return = accumulated_return(asset_return)
```
